````python
import matplotlib.pyplot as plt
import numpy as np
from VegasAfterglow import ISM, TophatJet, Observer, Radiation, Model
````

````python
medium = ISM(n_ism=1)

jet = TophatJet(theta_c=0.3, E_iso=1e52, Gamma0=100)

z = 0.1
obs = Observer(lumi_dist=1e26, z=z, theta_obs=0.)

rad = Radiation(eps_e=1e-1, eps_B=1e-3, p=2.3, ssc=True)

model = Model(jet=jet, medium=medium, observer=obs, fwd_rad=rad, resolutions=(0.15,1.0,10))
````

````python
details = model.details(1e0,1e10)

attrs =['Gamma', 'B_comv', 'N_p','r','N_e','I_nu_max']
ylabels = [r'$\Gamma$', r'$B^\prime$ [G]', r'$N_p$', r'$r$ [cm]', r'$N_e$', r'$I_{\nu, \rm max}^\prime$ [erg/s/Hz]']

frames = ['t_src', 't_comv', 't_obs']
titles = ['source frame', 'comoving frame', 'observer frame']
colors = ['C0', 'C1', 'C2']
xlabels = [r'$t_{\rm src}$ [s]', r'$t^\prime$ [s]', r'$t_{\rm obs}$ [s]']
plt.figure(figsize= (4.2*len(frames), 3*len(attrs)))

for i, frame in enumerate(frames):
    for j, attr in enumerate(attrs):
        plt.subplot(len(attrs), len(frames) , j * len(frames) + i + 1)
        if j == 0:
            plt.title(titles[i])
        value = getattr(details.fwd, attr)
        if frame == 't_src':
            t = getattr(details, frame)
        else:
            t = getattr(details.fwd, frame)
        plt.loglog(t[0, 0, :], value[0, 0, :], color='k',lw=2.5)
        plt.loglog(t[0, 0, :], value[0, 0, :], color=colors[i])

        plt.xlabel(xlabels[i])
        plt.ylabel(ylabels[j])

plt.tight_layout()
plt.savefig('shock_quantities.png', dpi=300,bbox_inches='tight')
````

## Per-Cell Spectrum Evaluation

The `details()` output provides callable accessors for evaluating comoving-frame spectra at arbitrary frequencies for each grid cell:
- `sync_spectrum[i,j,k](nu_comv)` — synchrotron specific intensity $I_\nu$ [erg/s/Hz/cm²/sr]
- `ssc_spectrum[i,j,k](nu_comv)` — SSC specific intensity (requires `ssc=True`)
- `Y_spectrum[i,j,k](gamma)` — Compton-Y parameter $Y(\gamma)$ (dimensionless)

````python
nu_comv = np.logspace(6, 22, 300)  # comoving frame frequency [Hz]
k = len(details.t_src[0, 0, :]) // 2  # pick a mid-time cell

# Synchrotron spectrum
I_syn = details.fwd.sync_spectrum[0, 0, k](nu_comv)

# SSC spectrum
I_ssc = details.fwd.ssc_spectrum[0, 0, k](nu_comv)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].loglog(nu_comv, I_syn)
axes[0].set_xlabel(r"$\nu'$ [Hz]")
axes[0].set_ylabel(r"$I_\nu'$ [erg/s/Hz/cm$^2$/sr]")
axes[0].set_title('Synchrotron Spectrum')

axes[1].loglog(nu_comv, I_ssc)
axes[1].set_xlabel(r"$\nu'$ [Hz]")
axes[1].set_ylabel(r"$I_\nu'$ [erg/s/Hz/cm$^2$/sr]")
axes[1].set_title('SSC Spectrum')

plt.tight_layout()
````

````python
# Compton-Y parameter as a function of electron Lorentz factor
gamma = np.logspace(1, 8, 200)
Y = details.fwd.Y_spectrum[0, 0, k](gamma)

plt.figure(figsize=(5, 4))
plt.loglog(gamma, Y)
plt.xlabel(r'$\gamma$')
plt.ylabel(r'$Y(\gamma)$')
plt.title('Compton-Y Parameter')
plt.tight_layout()
````
